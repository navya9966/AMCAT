### Times of India article dated Jan 18, 2019 states that â€œAfter doing your Computer Science Engineering if you take up jobs as a Programming Analyst, Software Engineer, Hardware Engineer and Associate Engineer you can earn up to 2.5-3 lakhs as a fresh graduate.â€


```python
from scipy import stats
relevant_roles = ['programmer Analyst', 'software engineer', 'hardware engineer', 'associate engineer']
filtered_df = df[df['Designation'].isin(relevant_roles)]
salary_data = filtered_df['Salary']
claimed_mean_salary = 2.75 * 100000 
t_stat, p_value = stats.ttest_1samp(salary_data, claimed_mean_salary)
print(f"Mean Salary of Selected Roles: {salary_data.mean():.2f}")
print(f"Claimed Mean Salary: {claimed_mean_salary:.2f}")
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")
alpha = 0.05 
if p_value < alpha:
 print("Reject the null hypothesis: The average salary is significantly different from the claimed mean")
else:
 print("Fail to reject the null hypothesis: There is no significant difference between the average salary")
```

### Is there a relationship between gender and specialization? (i.e. Does the preference of Specialisation depend on the Gender?)

```python
from scipy import stats as st
cont_table=pd.crosstab(index=df["Specialization"],columns=df["Gender"])
Chi2_stat,p_value,dof,exp_freq=st.chi2_contingency(cont_table)
alpha = 0.05 
if p_value < alpha:
 print("Reject the null hypothesis: There is a significant difference between the gender and Specialization")
else:
 print("Fail to reject the null hypothesis: There is no significant difference between the gender and Specialization")
```

### Conclusion
The analysis of the AMCAT dataset provides valuable insights into salary patterns, specialization, and the skills of recent
graduates across different roles. Below are the main findings:
### Salary Patterns
The statistical analysis shows that the average salaries for roles like Programming Analyst, Software Engineer, Hardware
Engineer, and Associate Engineer align with the figures reported in the Times of India article. There is no significant discrepancy
between the expected and actual salaries, supporting the industry standards for these positions.
### Impact of Specialization
Graduates specializing in Computer Science and IT-related fields tend to earn higher salaries, reflecting the growing demand for
these technical skills in the industry.
### Gender Distribution
The dataset highlights an uneven representation of male and female graduates in certain job roles, which may indicate gender
imbalances or disparities within particular specializations and positions.
### Skill Relevance
Technical skills, including programming and computer science, show a strong positive correlation with salary, underlining their
importance for higher earnings. Additionally, soft skills like conscientiousness, agreeableness, and openness to experience
demonstrate a moderate correlation with both job performance and salary, showing their relevance in the workplace.
### Educational Influence
Graduates from Tier 1 colleges tend to earn higher salaries compared to those from Tier 2 and Tier 3 institutions. This suggests
that the reputation of the educational institution plays a role in initial job placement and salary levels.
